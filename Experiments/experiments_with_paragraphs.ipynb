```python
from sentence_transformers import SentenceTransformer, util
model = SentenceTransformer('sentence-transformers/all-roberta-large-v1')

def sentence_similarity(text1, text2):
    sentences = [text1, text2]
    
    embedding_1= model.encode(sentences[0], convert_to_tensor=True)
    embedding_2 = model.encode(sentences[1], convert_to_tensor=True)

    return util.pytorch_cos_sim(embedding_1, embedding_2)[0][0].item()
```

```python
import warnings
warnings.filterwarnings("ignore", category=UserWarning)

import spacy
nlp = spacy.load("ru_core_news_sm")

def text_similarity(text1: str, text2: str):
    text1_nlp = nlp(text1)
    text2_nlp = nlp(text2) 

    return text1_nlp.similarity(text2_nlp)
```

```python
from math import*

def jaccard_similarity(x,y):

    intersection_cardinality = len(set.intersection(*[set(x), set(y)]))
    union_cardinality = len(set.union(*[set(x), set(y)]))
    return intersection_cardinality/float(union_cardinality)
```

```python
from sentence_transformers import SentenceTransformer, util
model = SentenceTransformer('ai-forever/sbert_large_nlu_ru')

def sentence_similarity_sber(text1, text2):
    sentences = [text1, text2]
    
    embedding_1= model.encode(sentences[0], convert_to_tensor=True)
    embedding_2 = model.encode(sentences[1], convert_to_tensor=True)

    return util.pytorch_cos_sim(embedding_1, embedding_2)[0][0].item()
```

```python
doc1 = 'Our main goal is find surprise'
doc2 = 'What a surprise you find'

# Tokenize the two documents
doc1_words = set(doc1.lower().split())
doc2_words = set(doc2.lower().split())

# Calculate the total number of unique words in both documents
total_words = len(doc1_words.union(doc2_words))

# Calculate the number of words that are in common between the two documents
common_words = len(doc1_words.intersection(doc2_words))

# Calculate the percentage of similar words
similarity_percentage = (common_words / total_words) * 100

# Print the percentage of similar words
print(similarity_percentage)
```
