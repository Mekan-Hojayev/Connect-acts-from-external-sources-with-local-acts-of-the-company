```python
documents = [text2[10:],
             text1[10:]]


#  ["Работодатели ведут документальный учет всех проведенных стажировок каждого водителя.",
#              "Работодатели ведут учет всех проведенных стажировок каждого водителя.",
#              "Стажировка и аттестация водителя проводится в связи с: трудоустройством в СП",
#              "Работодатели ведут документальный учет всех проведенных стажировок каждого работника."
#              ]


#     "That restaurant was not as good as the last movie I watched.",
#     "I'm selling a used car in good condition",
#     "Food was okay, the rest so so",
#     "I love cats, but don't really like hyenas",
#     "On the road, you must be careful",
# ]
vectors = [
  # tokenize the document, return it as PyTorch tensors (vectors),
  # and pass it onto the model
  model(**tokenizer(document, return_tensors='pt').to("cpu"))[0].detach().squeeze()
  for document in documents
]

[v.size() for v in vectors]
# => [torch.Size([15, 768]),
#     torch.Size([12, 768]),
#     torch.Size([10, 768]),
#     torch.Size([15, 768]),
#     torch.Size([10, 768])]
```

```python
vectors[0].shape[1]
```

```python
import torch

averaged_vectors = [torch.mean(vector, dim=0) for vector in vectors]

[v.size() for v in averaged_vectors]
# => [torch.Size([768]), torch.Size([768]), torch.Size([768]), torch.Size([768]), torch.Size([768])]
```

```python
for vecs in averaged_vectors:
  vecs.to("cpu")
```

```python
def encode(document: str) -> torch.Tensor:
  tokens = tokenizer(document, return_tensors='pt')
  vector = model(**tokens)[0].detach().squeeze()
  return torch.mean(vector, dim=0)
```

```python
import faiss
import numpy as np

index = faiss.IndexIDMap(faiss.IndexFlatIP(vectors[0].shape[1])) # the size of our vector space
# index all the documents, we need them as numpy arrays first
index.add_with_ids(
    np.array([t.cpu().numpy() for t in averaged_vectors]),
    # the IDs will be 0 to len(documents)
    np.array(range(0, len(documents)))
)

def search(query: str, k=1):
  encoded_query = encode(query).unsqueeze(dim=0).numpy()
  top_k = index.search(encoded_query, k)
  # print(top_k[0][0][0])
  scores = top_k[0][0] / top_k[0][0][0]
  results = [documents[_id] for _id in top_k[1][0]]
  # print(query)
  return query, list(zip(results[1:], scores[1:]))
```

```python
import torch
import pandas as pd
from tqdm.autonotebook import tqdm
from transformers import T5Tokenizer, T5ForConditionalGeneration
from transformers import AutoTokenizer
```

```python
MODEL_NAME = 'UrukHan/t5-russian-spell' # 'sberbank-ai/sbert_large_mt_nlu_ru' # 'UrukHan/t5-russian-spell'

tokenizer = AutoTokenizer.from_pretrained(MODEL_NAME)

model = T5ForConditionalGeneration.from_pretrained(MODEL_NAME)
device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')
model.to(device)
```

```python
myfile = open('/content/drive/MyDrive/Track Track/0001202211240047.pdf', mode='rb')
pdf_document_file = PyPDF2.PdfReader(myfile)

file2 = []
for i in range(len(pdf_document_file.pages)):
  pages = pdf_document_file.pages[i]
  pages = pages.extract_text().replace("\n", " ")
  file2.append(pages)

test_inputs = []

for i in tqdm(range(len(file2))):
    toxic_test = file2[i]

    test_inputs.append(toxic_test)

test_input_ids = tokenizer.batch_encode_plus(test_inputs, padding='max_length', max_length=256, truncation=True, return_tensors='pt')['input_ids']
test_attention_mask = test_input_ids.gt(0).long()
```

```python
model.eval()
generated_text2 = []
batch_size = 1
num_batches = (len(test_inputs) + batch_size - 1) // batch_size

for i in tqdm(range(num_batches)):
    start_idx = i * batch_size
    end_idx = min((i + 1) * batch_size, len(test_inputs))
    input_ids = test_input_ids[start_idx:end_idx].to(device)
    attention_mask = test_attention_mask[start_idx:end_idx].to(device)
    outputs = model.generate(input_ids=input_ids, attention_mask=attention_mask, max_length=256, num_beams=8, early_stopping=True)
    for output in outputs:
        title = tokenizer.decode(output, skip_special_tokens=True)
        generated_text2.append(title)
```
